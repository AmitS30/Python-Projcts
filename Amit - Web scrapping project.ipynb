```python
v=soup.find(id='makeTabsContentWrapper')
print(v)
```

```python
list_items=v.find(class_="listitems")
print(list_items)
```

```python
model=list_items.find(class_="front").get_text()
print(model)
```

```python
img=list_items("img")
print(img)
```

```python
models=list_items.find(class_="bikeTitle margin-bottom10").get_text()
models
```

```python
models1=v.find_all(class_="bikeTitle margin-bottom10")
models1
```

```python
name = [i.get_text() for i in models1]
name
```

```python
details=list_items(class_="text-xt-light-grey font14 margin-bottom15")
details
```

```python
prize=list_items(class_="text-bold")
prize
```

```python
prize_list=[i.get_text() for i in prize]
prize_list
```

```python
details_list=[i.get_text() for i in details]
details_list
```

```python
data={"name":name,"features":details_list,"prize":prize_list}
```
